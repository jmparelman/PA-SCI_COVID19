```python
import pandas as pd
```

```python
fpath='../data/DOC_dashboard_export/DOC-COVID-19-Dashboard-Data-Mon_Jul_19_14:51:23_EDT_2021.xlsx'
```

```python
# load all sheets
doc_data = pd.read_excel(fpath, sheet_name=None )
```

```python
# number of sheets
len(doc_data)
```

```python
# sheet names
doc_data.keys()
```

```python
cases_df = doc_data['Inmate New Cases History']
```

```python
cases_df
```

```python
cases_long_df=cases_df.drop(columns='TOTAL:').melt(id_vars='Date', 
                                     var_name='SCI',
                                     value_name='incarcerated_person_positive')

cases_long_df.rename(columns = {'Date': 'date'}, inplace=True)
cases_long_df
```

```python
cases_long_df.to_csv('../data/latest_data/PA_DOC_dashboard_cases.csv', index=False)
```
