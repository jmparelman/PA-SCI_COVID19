```python
import pandas as pd
```

```python
fpath='../data/DOC_dashboard_export/DOC-COVID-19-Dashboard-Data-Tue_Dec_28_10:15:06_EST_2021.xlsx'
```

```python
# load all sheets
doc_data = pd.read_excel(fpath, sheet_name=None )
```

```python
# number of sheets
len(doc_data)
```

```python
# sheet names
doc_data.keys()
```

```python
cases_df = doc_data['Inmate New Cases History']
```

```python
cases_df
```

```python
cases_long_df=cases_df.drop(columns='TOTAL:').melt(id_vars='Date', 
                                     var_name='SCI',
                                     value_name='incarcerated_person_positive')

cases_long_df.rename(columns = {'Date': 'date'}, inplace=True)
cases_long_df
```

```python
cases_long_df.to_csv('../data/latest_data/PA_DOC_dashboard_cases.csv', index=False)
```

```python
doc_data['Inmate']
```

```python
cols_to_use = ['SCI', 'POPULATION','ACTIVE POSITIVE CASES', 'CUMULATIVE CASES', 'DEATHS']


cmap = {'POPULATION': 'population',
        'ACTIVE POSITIVE CASES': 'ip_active_cases',
        'CUMULATIVE CASES': 'ip_cases',
        'DEATHS': 'ip_deaths'
       }

ip_df=doc_data['Inmate'][cols_to_use].rename(columns = cmap)
```

```python
scols_to_use = ['SCI', 'ACTIVE CASES', 'CUMULATIVE CASES']

scmap = {
            'ACTIVE CASES': 'staff_active_cases',
            'CUMULATIVE CASES': 'staff_cases'
        }

staff_df=doc_data['SCI Staff'][scols_to_use].rename(columns=scmap)
```

```python
latest_df = pd.merge(ip_df, staff_df).drop(25)
latest_df['SCI']=latest_df['SCI'].str.title()
```

```python
latest_df.to_csv('../data/latest_data/PA_DOC_dashboard_latest.csv', index=False)
```
